```python
def extract_utm(list_of_tags, tag_name):
    try:
        tag_value = [x for x in list_of_tags if x['name'] == tag_name][0]['values'][0]['value']
    except:
        tag_value = None
    return tag_value
```

```python
def get_credentials(prefix, param_list):
    ssm_parameters = ssm.get_parameters(
        Names=[prefix + p for p in param_list], WithDecryption=True)
    return {x['Name'].replace(prefix, ''): x['Value'] for x in ssm_parameters['Parameters']}
```

```python
def get_data(base_url, endpoint, access_token, date_from_str=None):
    data = []
    method = 'GET'
    ids = set()
    i = 0

    headers = {
        'User-Agent': 'amoCRM-oAuth-client/1.0',
        'Authorization': 'Bearer {}'.format(access_token)
    }

    if date_from_str != None:
        headers['IF-MODIFIED-SINCE'] = date_from_str

    while True:
        print(i)
        params = {
            'limit_rows': 500,
            'limit_offset': i
        }
        res = requests.get(url=base_url+endpoint,
                           params=params, headers=headers)
        try:
            new = json.loads(res.text)
            for x in new['_embedded']['items']:
                if x['id'] not in ids:
                    ids.add(x['id'])
                    data.append(x)
            i += 500
        except:
            print('Done')
            break
    return data
```

```python
def get_pipeline_stages(base_url, access_token):
    endpoint = '/api/v2/pipelines'
    method = 'GET'

    headers = {
        'User-Agent': 'amoCRM-oAuth-client/1.0',
        'Authorization': 'Bearer {}'.format(access_token)
    }

    res = requests.get(url=base_url+endpoint, headers=headers)
    res_text = json.loads(res.text)

    stages = []

    for pipeline in res_text['_embedded']['items'].values():
        pl_stages = list(pipeline['statuses'].values())

        for p in pl_stages:
            p.update({'pipeline_id': pipeline['id'], 'pipeline_name': pipeline['name']})
            p.pop('color')
        stages += pl_stages
        
    return stages
```

```python
def transform_leads(clean_leads):
    data = []
    for l in clean_leads:
        data.append({
            'lead_id': l['id'],
            'lead_name': l['name'],
            'lead_created_at': l['created_at'],
            'lead_closed_at': l['closed_at'],
            'lead_sale': l['sale'],
            'lead_closest_task_at': l['closest_task_at'],
            'lead_loss_reason_id': l['loss_reason_id'],
            'lead_pipeline_id': l['pipeline_id'],
            'lead_status_id': l['status_id'],
            'lead_is_deleted': l['is_deleted'],
            'lead_utm_source':   extract_utm(l['custom_fields'], 'utm_source'),
            'lead_utm_medium':   extract_utm(l['custom_fields'], 'utm_campaign'),
            'lead_utm_campaign': extract_utm(l['custom_fields'], 'utm_medium'),
            'lead_main_contact': l['main_contact'].get('id'),
            'lead_tags': ' ^ '.join([t.get('name') for t in l['tags']])
        }
    )
    return data
```

```python
def transform_contacts(contacts):
    data = []
    for c in contacts:
        try:
            email = [x for x in c['custom_fields'] if x['code']
                     == 'EMAIL'][0]['values'][0]['value']
        except:
            email = None

        try:
            phone = [x for x in c['custom_fields'] if x['code']
                     == 'PHONE'][0]['values'][0]['value']
        except:
            phone = None

        data.append({
            'contact_id': c['id'],
            'contact_name': c['name'],
            'contact_responsible_user_id': c['responsible_user_id'],
            'contact_created_by': c['created_by'],
            'contact_created_at': c['created_at'],
            'contact_email': email,
            'contact_phone': phone
        })
    return data
```
