```python
import requests
import json
import os
from datetime import datetime, date, timedelta

from urllib.parse import urlparse, parse_qs, quote, urlencode

import boto3
import pandas as pd
```

```python
ssm = boto3.client('ssm')
s3 = boto3.resource('s3')
```

```python
def get_credentials(prefix, param_list):
    ssm_parameters = ssm.get_parameters(
        Names=[prefix + p for p in param_list], WithDecryption=True)
    return {x['Name'].replace(prefix, ''): x['Value'] for x in ssm_parameters['Parameters']}


def set_credentials(prefix, param_name, value):
    path = prefix + param_name
    return ssm.put_parameter(Name=path, Value=str(value), Overwrite=True, Type='SecureString')


def extract_utm(list_of_tags, tag_name):
    try:
        tag_value = [x for x in list_of_tags if x['name']
                     == tag_name][0]['values'][0]['value']
    except:
        tag_value = None
    return tag_value


def refresh_tokens(base_url, client_id, client_secret, refresh_token):
    """Refresh access and refresh tokens in Amocrm"""

    endpoint = '/oauth2/access_token'
    method = 'POST'

    headers = {
        'User-Agent': 'amoCRM-oAuth-client/1.0'
    }

    params = {
        'client_id': client_id,
        'client_secret': client_secret,
        'grant_type': 'refresh_token',
        'refresh_token': refresh_token,
        'redirect_uri': 'https://smysl.io'
    }
    res = requests.post(url=base_url+endpoint, data=params, headers=headers)
    body = json.loads(res.text)

    return body

def get_funnel_stages(base_url, access_token, funnel_id):
    endpoint = '/api/v2/pipelines'
    method = 'GET'

    headers = {
        'User-Agent': 'amoCRM-oAuth-client/1.0',
        'Authorization': 'Bearer {}'.format(access_token)
    }

    res = requests.get(url=base_url+endpoint, headers=headers)
    funnels = json.loads(res.text)
    return funnels['_embedded']['items'][funnel_id]['statuses']
```

```python
def get_pipeline_stages(base_url, access_token):
    endpoint = '/api/v2/pipelines'
    method = 'GET'

    headers = {
        'User-Agent': 'amoCRM-oAuth-client/1.0',
        'Authorization': 'Bearer {}'.format(access_token)
    }

    res = requests.get(url=base_url+endpoint, headers=headers)
    res_text = json.loads(res.text)

    stages = []

    for pipeline in res_text['_embedded']['items'].values():
        pl_stages = list(pipeline['statuses'].values())

        for p in pl_stages:
            p.update({'pipeline_id': pipeline['id'], 'pipeline_name': pipeline['name']})
            p.pop('color')
        stages += pl_stages
        
    return stages
```

```python
pipeline_stages = get_pipeline_stages(base_url, access_token)
stages_df = pd.DataFrame(pipeline_stages)
pipeline_stages_df.columns = ['stage_id', 'stage_name', 'stage_sort', 'stage_is_editable', 'pipeline_id', 'pipeline_name']
```

```python
pipelines = json.loads(res.text)
```

```python
pl = {p[1]['name']:p[1]['id'] for p in pipelines['_embedded']['items'].items()}
```

```python

events = []
method = 'GET'
ids = set()
i = 0

endpoint = '/api/v4/events'

headers = {
    'User-Agent': 'amoCRM-oAuth-client/1.0',
    'Authorization': 'Bearer {}'.format(access_token)
}

```

```python
step = 100

while True:

    params = {
        'limit_rows': step,
        'limit_offset': i,
        'filter[entity]': 'lead',
        'filter[type]': ['lead_added', 'lead_status_changed']
    }

    res = requests.get(url=base_url+endpoint,
                       params=params, headers=headers)
    print(i, res.status_code, len(events))
    
    try:
        new = json.loads(res.text)
        events += new['_embedded']['events']
        i += step
    except:
        print('Done')
```

```python
clean_events = []
error_events = []
for x in events:
    try:
        clean_events.append({
            'type': x['type'],
            'entity_id': x['entity_id'],
            'created_at': x['created_at'],
            'old_pipeline': x['value_before'][0]['lead_status']['pipeline_id'],
            'new_pipeline': x['value_after'][0]['lead_status']['pipeline_id'],
            'old_status': x['value_before'][0]['lead_status']['id'],
            'new_status': x['value_after'][0]['lead_status']['id']
        })
    except:
        error_events.append(x)
```

```python
# когорты
pipeline_name = 'Обращения в ОП'

pl[pipeline_name]
```

```python
df = pd.DataFrame(clean_events)
df = df.query('new_pipeline == @pl[@pipeline_name]')
df.created_at = pd.to_datetime(df.created_at, unit='s')
```

```python
created_week = (df.created_at - pd.to_timedelta(df.created_at.dt.dayofweek, unit='d')).dt.date
```

```python
created_week.name = 'created_week'
```

```python
df = df.join(created_week)
```

```python
# reg_date

new_leads = df.groupby('entity_id')['created_week'].min()
new_leads = new_leads.reset_index()
new_leads.head()
```

```python
target_stage_name = 'выставлен счет'

mask = pipeline_stages_df['pipeline_id'] == pl[pipeline_name]
mask = pipeline_stages_df['stage_name'] == target_stage_name
target_stage_id = int(pipeline_stages_df[mask]['stage_id'])
```

```python

converted = (
    df[df['new_status'] == target_stage_id]
    .groupby('entity_id')['created_week'].min()
)
converted = converted.reset_index()
converted.columns = ['entity_id', 'converted_week']
```

```python
merged = new_leads.merge(converted, on='entity_id', how='left')
```
